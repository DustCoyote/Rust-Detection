```python
import os
import glob
import random
import numpy as np
from PIL import Image, ImageOps

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, random_split
import torchvision.transforms as T

import matplotlib.pyplot as plt

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
```

```python
class RustSegDataset(Dataset):
    def __init__(self, images_dir, masks_dir, img_size=(256, 256), augment=False):
        self.images_dir = images_dir
        self.masks_dir  = masks_dir
        self.img_size   = img_size
        self.augment    = augment

        self.fnames = sorted([
            os.path.splitext(f)[0]
            for f in os.listdir(images_dir)
            if f.lower().endswith((".png", ".jpg", ".jpeg"))
        ])

        self.to_tensor_img = T.Compose([
            T.Resize(img_size),
            T.ToTensor(),
        ])

        self.to_tensor_mask = T.Compose([
            T.Resize(img_size, interpolation=Image.NEAREST),
            T.ToTensor(),
        ])

    def __len__(self):
        return len(self.fnames)

    def __getitem__(self, idx):
        fname = self.fnames[idx]
        img_path  = os.path.join(self.images_dir, fname + ".png")
        mask_path = os.path.join(self.masks_dir,  fname + ".png")

        image = Image.open(img_path).convert("RGB")
        mask  = Image.open(mask_path).convert("L")

        if self.augment and random.random() < 0.5:
            image = ImageOps.mirror(image)
            mask  = ImageOps.mirror(mask)

        image = self.to_tensor_img(image)
        mask  = self.to_tensor_mask(mask)
        mask = (mask > 0.5).float()

        return image, mask
```

```python
images_dir = "dataset/content/dataset/images"
masks_dir  = "dataset/content/dataset/masks"

dataset = RustSegDataset(images_dir, masks_dir, img_size=(256,256), augment=True)
print("????:", len(dataset))

# 8:2 split train/val
val_ratio = 0.2
val_len = int(len(dataset) * val_ratio)
train_len = len(dataset) - val_len

train_dataset, val_dataset = random_split(dataset, [train_len, val_len])
print("Train:", len(train_dataset), "Val:", len(val_dataset))

batch_size = 4
num_workers = 0 if os.name == "nt" else 2
pin_memory = torch.cuda.is_available()

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=num_workers, pin_memory=pin_memory)
val_loader   = DataLoader(val_dataset,   batch_size=batch_size, shuffle=False, num_workers=num_workers, pin_memory=pin_memory)
```

```python
class DoubleConv(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, padding=1),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, padding=1),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True),
        )

    def forward(self, x):
        return self.conv(x)

class UNet(nn.Module):
    def __init__(self, n_channels=3, n_classes=1):
        super().__init__()

        self.down1 = DoubleConv(n_channels, 64)
        self.down2 = DoubleConv(64, 128)
        self.down3 = DoubleConv(128, 256)
        self.down4 = DoubleConv(256, 512)

        self.pool = nn.MaxPool2d(2)

        self.bottleneck = DoubleConv(512, 1024)

        self.up4 = nn.ConvTranspose2d(1024, 512, kernel_size=2, stride=2)
        self.conv4 = DoubleConv(1024, 512)

        self.up3 = nn.ConvTranspose2d(512, 256, kernel_size=2, stride=2)
        self.conv3 = DoubleConv(512, 256)

        self.up2 = nn.ConvTranspose2d(256, 128, kernel_size=2, stride=2)
        self.conv2 = DoubleConv(256, 128)

        self.up1 = nn.ConvTranspose2d(128, 64, kernel_size=2, stride=2)
        self.conv1 = DoubleConv(128, 64)

        self.out_conv = nn.Conv2d(64, n_classes, kernel_size=1)

    def forward(self, x):
        c1 = self.down1(x)
        p1 = self.pool(c1)

        c2 = self.down2(p1)
        p2 = self.pool(c2)

        c3 = self.down3(p2)
        p3 = self.pool(c3)

        c4 = self.down4(p3)
        p4 = self.pool(c4)

        bottleneck = self.bottleneck(p4)

        u4 = self.up4(bottleneck)
        u4 = torch.cat([u4, c4], dim=1)
        c4 = self.conv4(u4)

        u3 = self.up3(c4)
        u3 = torch.cat([u3, c3], dim=1)
        c3 = self.conv3(u3)

        u2 = self.up2(c3)
        u2 = torch.cat([u2, c2], dim=1)
        c2 = self.conv2(u2)

        u1 = self.up1(c2)
        u1 = torch.cat([u1, c1], dim=1)
        c1 = self.conv1(u1)

        out = self.out_conv(c1)  # [B,1,H,W]
        return out
```

```python
def dice_coeff(pred, target, eps=1e-7):
    # pred: logits ??probs，這�??��???probs
    pred = torch.sigmoid(pred)
    pred = (pred > 0.5).float()

    intersection = (pred * target).sum(dim=(1,2,3))
    union = pred.sum(dim=(1,2,3)) + target.sum(dim=(1,2,3))
    dice = (2 * intersection + eps) / (union + eps)
    return dice.mean()

def iou_score(pred, target, eps=1e-7):
    pred = torch.sigmoid(pred)
    pred = (pred > 0.5).float()

    intersection = (pred * target).sum(dim=(1,2,3))
    union = pred.sum(dim=(1,2,3)) + target.sum(dim=(1,2,3)) - intersection
    iou = (intersection + eps) / (union + eps)
    return iou.mean()

model = UNet(n_channels=3, n_classes=1).to(device)

bce_loss = nn.BCEWithLogitsLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
```

```python
def train_one_epoch(model, loader, optimizer):
    model.train()
    total_loss = 0
    total_dice = 0
    total_iou  = 0
    for images, masks in loader:
        images = images.to(device)
        masks  = masks.to(device)

        optimizer.zero_grad()
        logits = model(images)
        loss = bce_loss(logits, masks)

        loss.backward()
        optimizer.step()

        total_loss += loss.item() * images.size(0)
        total_dice += dice_coeff(logits, masks).item() * images.size(0)
        total_iou  += iou_score(logits, masks).item() * images.size(0)

    n = len(loader.dataset)
    return total_loss/n, total_dice/n, total_iou/n

def eval_one_epoch(model, loader):
    model.eval()
    total_loss = 0
    total_dice = 0
    total_iou  = 0
    with torch.no_grad():
        for images, masks in loader:
            images = images.to(device)
            masks  = masks.to(device)

            logits = model(images)
            loss = bce_loss(logits, masks)

            total_loss += loss.item() * images.size(0)
            total_dice += dice_coeff(logits, masks).item() * images.size(0)
            total_iou  += iou_score(logits, masks).item() * images.size(0)

    n = len(loader.dataset)
    return total_loss/n, total_dice/n, total_iou/n
```

```python
num_epochs = 30
best_val_dice = 0.0

for epoch in range(1, num_epochs+1):
    train_loss, train_dice, train_iou = train_one_epoch(model, train_loader, optimizer)
    val_loss,   val_dice,   val_iou   = eval_one_epoch(model, val_loader)

    print(f"[Epoch {epoch:02d}] "
          f"Train Loss: {train_loss:.4f}  Dice: {train_dice:.4f}  IoU: {train_iou:.4f} | "
          f"Val Loss: {val_loss:.4f}  Dice: {val_dice:.4f}  IoU: {val_iou:.4f}")

    # ?��??�佳模?��?�?validation Dice ?��?）
    if val_dice > best_val_dice:
        best_val_dice = val_dice
        torch.save(model.state_dict(), "best_unet_rust.pth")
        print(f"  ?�� New best model saved! Val Dice = {best_val_dice:.4f}")
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import os
import random
from PIL import Image

# ?? device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model_path = "best_unet_rust.pth"
if not os.path.exists(model_path):
    raise FileNotFoundError("??? best_unet_rust.pth??????????????")

# ????
model.load_state_dict(torch.load(model_path, map_location=device))
model.to(device)
model.eval()

# ======== ?? test ????????? ========
img_dir = "test"
files = [f for f in os.listdir(img_dir) if f.lower().endswith((".jpg", ".png", ".jpeg"))]

if len(files) == 0:
    raise ValueError("?????? jpg/png ??")

random_file = random.choice(files)
img_path = os.path.join(img_dir, random_file)

print("?????????:", img_path)

# ======== ?? ========
img_pil = Image.open(img_path).convert("RGB")

# Resize ???????? (256x256)
TARGET_SIZE = (256, 256)
img_rgb_resized = np.array(img_pil.resize(TARGET_SIZE, Image.BILINEAR))

# ?? tensor
img_tensor = torch.tensor(img_rgb_resized / 255.0, dtype=torch.float32)
img_tensor = img_tensor.permute(2, 0, 1).unsqueeze(0).to(device)

# ======== ?? ========
with torch.no_grad():
    logits = model(img_tensor)
    probs = torch.sigmoid(logits)
    pred = (probs > 0.5).float()

pred_np = pred[0].cpu().squeeze().numpy()

# ======== ?? overlay ========
overlay = img_rgb_resized.copy()
alpha = 0.5

red_mask = np.zeros_like(overlay)
red_mask[:, :, 0] = 255

overlay[pred_np > 0.5] = (
    overlay[pred_np > 0.5] * (1 - alpha) +
    red_mask[pred_np > 0.5] * alpha
).astype(np.uint8)

# ======== ?? ========
plt.figure(figsize=(18,6))

plt.subplot(1,3,1)
plt.title(f"Random Image
{random_file}")
plt.imshow(img_rgb_resized)
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Pred Mask")
plt.imshow(pred_np, cmap='gray')
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Overlay")
plt.imshow(overlay)
plt.axis("off")

plt.show()
```
